Tensor 是用來表示向量、矩陣或其他維度資料的一種資料結構及容器。
如何得知張量的維度（amount of axis）？ 可尤其索引值的 **個數** 得知。

## Session 3.2 Tensor: 多維陣列

```python
'''
以下在比較array 跟tensor
'''
a=[1.0,2.0,1.0]
print(a[0])
a[0]=3.0
print(a)
```

```python
import torch
a=torch.ones(3) #創造一個3維的1軸向量，其元素為1
print(a)
print(a[1])
print(float(a[1]))
```

```python
a[2]=2.0
a
```

```python
points=torch.zeros(6)
points[0]=4.0
points[1]=1.0
points[2]=5.0
points[3]=3.0
points[4]=2.0
points[5]=1.0
points

#與以下等價
points2=torch.tensor([4.0,1.0,5.0,3.0,2.0,1.0])
points2

#提取A座標
float(points[0]),float(points[1])
```

```python
## 創造2軸張量
points=torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,1.0]])
points[0]
```

```python
points.shape
#代表points是一個2軸張量（第0軸有3維，第1軸有2維），.shape用來回傳每個軸的維度
```

```python
points=torch.zeros(3,2)
points #透過指定shape來建立tensor
```

```python
#提取2軸tensor的個別元素（by 2 index）
points=torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,1.0]])
points
```
