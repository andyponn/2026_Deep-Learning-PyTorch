## Session 3.4 為張量命名

利用names參數替每個軸命名（typr: string）

```python
import torch
weights_named=torch.tensor([0.2126, 0.7152,0.0722], names=["channels"])
weights_named
#refine_names()重設名稱
```

```python
img_t = torch.randn(3,5,5) #創造一個3軸的張量
batch_t = torch.randn(2,3,5,5) #創造一個4軸的張量
img_named = img_t.refine_names(..., 'channels','rows','columns') ## ...表示忽略前面，只幫最後三軸命名
batch_named = batch_t.refine_names(...,'channels','rows','columns')
print("img_named:" , img_t.shape, img_named.names)
print("batch_named:" , batch_t.shape, batch_named.names)
```

```python
#align_as()將weights_named內tensor的軸按照img_named順序進行排序，（會自動添加缺失的軸）
weights_aligned = weights_named.align_as(img_named) #weights_named為一軸張量
weights_aligned.shape, weights_aligned.names
#weights_aligned
```

```python
img_named
```

```python
weights_aligned
```

```python
gray_named = (img_named * weights_aligned).sum("channels")
gray_named.shape, gray_named.names
```

```python
gray_named
```

```python
## 有些函式不支援以張量維度名字來操作，重新命名為none
gray_plain = gray_named.rename(None)
gray_plain.shape, gray_plain.names
```

## Session 3.8 大小(size)、偏移(offset)及步長(stride)

* size(在numpy中稱作shape)表示每個軸上有多少元素<br>
* offset是張量首個元素在storage中的索引值（預設為0）<br>
* stride(為tuple——有序、不可變動的資料結構)為提取各軸下一個元素時，需要跳過的元素數量<br>
* 要知道索引為(i,j)的元素在storage的索引值，可透過 **storage_offset+stride[0]\*i +stride[1]\*j**

```python
points=torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,2.0]])
points.storage()
```

```python
secoond_point = points[1]
secoond_point.storage() # 可以發現與points張量指向統一個storage
```

```python
secoond_point.storage_offset()
## 由於跳過4.0, 1.0兩個元素，故offset=2
```

```python
secoond_point.size()
secoond_point.shape
```

```python
points.stride()
```

```python
secoond_point = points[1]
secoond_point.size()
```

```python
#因為輸出points second_point張量的storage為一致的，所以修改second_point的值，points的值也會被修改！
points = torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,1.0]])
second_point = points[1]
#second_point = points[1].clone()
## 解法：為了避免修改到母tensor的值，我們可以利用clone()將母張量所需的元素複製出來使用
print(second_point)
second_point[0]=10.0
points
```

```python
## transpose(把原本張量的stride元素對調，進而產生一個新tensor，不會佔用新的記憶體空間) of tensor
points = torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,1.0]])
print(points)
points_t=points.t()
points_t
```

```python
#check 2 tensor share common storage or not
print(id(points.storage()) == id(points_t.storage()))
#shape equal ??
print(points.stride())
print(points_t.stride())
```

```python
## 多軸張量的transpose
some_t =torch.ones(3,4,5)
#some_t = torch.arange(1, 61).reshape(3, 4, 5)
print(some_t)
print(some_t.shape)
transpose_t = some_t.transpose(0,2) #第0軸第2軸進行轉置
print(transpose_t.shape)
```

```python
print(some_t)
transpose_t = some_t.transpose(0,2) #第0軸第2軸進行轉置
print(transpose_t)
print(some_t.stride(), transpose_t.stride())
print("可以發現stride值對調了")
```

```python
## 有一些操作只能在連續張量進行，例如：view。故解法為： contiguous()
points.is_contiguous() # check
points_t.is_contiguous()
```

```python
# contiguous() can make a tensor change from non-continuous into a new continuous ones
# content is same but the storage will change
points = torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,1.0]])
print(points)
points_t=points.t()
print(points_t)
print(points_t.storage())
```

```python
print(points.stride())
print(points_t.stride())
print(points_t.storage())
points_t_cont = points_t.contiguous()
print(points_t_cont)
print(points_t_cont.stride()) #資訊改變stride
print(points_t_cont.storage()) #storage重新排列
```
