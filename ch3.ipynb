## Session 3.3 利用索引值操作張量

本節要介紹**範圍式索引法**（用於提取tensor中，一定範圍的元素），類似python list操作方法。<br>
**進階索引**(另一種強大的索引方式)將在chapter4中提到。

```python
some_list =list(range(6)) #產生一個從0到5的數字串列
some_list[:]              #提取串列中所有元素
```

```python
## Python list寫法
some_list[0:4] #提取no.0~no.3個元素
some_list[1:] #提取no.1~最後一個元素
some_list[:4] #提取no.0~no.3個元素
some_list[:-1] #提取no.0~倒數第二個元素
some_list[1:4:2] #提取no.1~no.3個元素，每次間隔2個元素
```

```python
## 對tensor使用範圍式索引法
points=torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,1.0]])
points
```

```python
points[1:,:]
```

```python
points[1:,0]
```
